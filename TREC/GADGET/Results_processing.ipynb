```python
import os
import re
import pandas as pd
import numpy as np
import collections
import math
```

```python
modelpath = 'GADGET/data/datasets/'
testpath = 'GADGET/data/test/datasets/'
```

```python
def resultprocess(textfile, foldername):
    with open(testpath+'/'+textfile, 'r') as fp:
        tweets = fp.read().splitlines()
    classes = os.listdir(modelpath+'/'+foldername)
    alltweets = []
    print("Total number of samples: ", len(tweets))
    for index, eachtweet in enumerate(tweets):
        tweetcontent = eachtweet.split()
        hashmap = collections.defaultdict()
        print("Finished tweets number: ", index)
        for each in classes:
            wt, norm = 0, 0
            with open(modelpath+'/'+foldername+'/'+each+'/run0/global_0.dat', 'r') as fp:
                content = fp.read().split()
            for i in range(min(len(content), len(tweetcontent))):
                wt += float(content[i].split(':')[1]) * float(tweetcontent[i].split(':')[1])
                norm += float(content[i].split(':')[1]) * float(content[i].split(':')[1])
            score = wt/norm
            score = 1 / (1 + math.exp(-score))
            hashmap[each] = score
        print("------------------------------------------------------")
        alltweets.append(hashmap)
    return alltweets
```

### Format Results

```python
hashmap= {
'floodChoco2019' : 'TRECIS-CTIT-H-Test-022', 
'fireAndover2019': 'TRECIS-CTIT-H-Test-023',
'earthquakeCalifornia2014': 'TRECIS-CTIT-H-Test-024',
'earthquakeBohol2013': 'TRECIS-CTIT-H-Test-025',
'hurricaneFlorence2018' : 'TRECIS-CTIT-H-Test-026',
'shootingDallas2017': 'TRECIS-CTIT-H-Test-027' ,
'fireYMM2016': 'TRECIS-CTIT-H-Test-028'}
```

```python
fullpath = '/home/harsha/TREC_Proj/testFeatures/'
files = os.listdir(fullpath)
```

```python
newhashmap = collections.defaultdict()
for file in files:
    df = pd.read_csv(fullpath+file).drop('Unnamed: 0', axis=1)
    newhashmap[df['topics'].tolist()[0]] = df['tweetids'].tolist()
```

### Normalize

```python
allfiles = os.listdir(testpath)
```

```python
allfiles = [each for each in allfiles if '.csv' in each]
```

```python
allfiles
```

```python
newhashmap.keys()
```

```python
len(newhashmap['earthquakeBohol2013'])
```

```python
sigmoid = lambda x: 1 / (1 + math.exp(-x))
for filename in allfiles:
    curfilename = filename.split('_')[0]
    
    df = pd.read_csv(testpath+'/'+filename).drop('Unnamed: 0', axis=1)
    
    columns = df.iloc[0].tolist()

    df.columns = columns

    df = df.drop(0, axis=0)
    
    for each in columns:
        df[each] = pd.to_numeric(df[each])
        df[each] = df[each].replace(to_replace=0, method='ffill')
        df[each] = df[each].apply(sigmoid)

    df['Threshold'] = df.sum(axis=1) / len(df.columns)

    results = lambda x: list(filter(None, x))

    df['results'] = list(np.where(df.gt(df['Threshold'],0), list(df.columns), ''))

    df['results'] = df['results'].apply(results)
        
    df['tweetids'] = newhashmap[curfilename]
    df['num'] = [hashmap[curfilename]]*len(df)
    df['buffer'] = ['Q0']*len(df)
    df['Identifier'] = ['run1']*len(df)
    df.to_csv(testpath+'/norm_'+filename)
```

### Write in text file

```python
alf = os.listdir(testpath)
```

```python
norms = [each for each in alf if 'norm' in each]
```

```python
norms[0]
```

```python
for each in norms:
    with open('results_'+each+'.txt','w') as fp:
        df = pd.read_csv(testpath+each)
        a, b, c, d, e, f, g = df['num'].tolist(), df['buffer'].tolist(), df['tweetids'].tolist(), list(df.index),\
        df['Threshold'].tolist(), df['results'].tolist(), df['Identifier'].tolist()
        for i,j,k,l,m,n,o in zip(a,b,c,d,e,f,g):
            fp.write("%s\t%s\t%s\t%s\t%s\t%s\t%s\n\n\n" %(i,j,k,l,m,n,o))
```
