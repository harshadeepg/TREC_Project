### F1 score

```python
def postprocess(path):
    events = os.listdir(path)
    error = []
    for anevent in events:
        classes = os.listdir(path+anevent)
        
        for aclass in classes:
            files = os.listdir(path+anevent+'/'+aclass+'/run0')
            files = [f for f in files if '.csv' in f]
            accuracy, obj_value_diff = [], []
            print("Writing for %s of %s " %(anevent, aclass) )
            for node in files:
                try:
                    df = pd.read_csv(path+anevent+'/'+aclass+'/run0/'+node)
                    accuracy.append(df['accuracy'].tolist()[-1])
                    obj_value_diff.append(df['obj_value_difference'].tolist()[-1])
                except Exception:
                    print("Error in reading: ", node)
                    error.append(path+anevent+'/'+aclass+'/run0/'+node)

            createdir = 'Results/'+anevent+'/'+aclass+'/' 
            os.makedirs(createdir)
            with open(createdir+'/result.txt', 'w') as fp:
                fp.write("%s\t%s\n%s\t%s" %('F1 score', np.mean(accuracy), 'Obj_value_diff', np.mean(obj_value_diff)))
                
            print("Finished writing for %s of %s " %(anevent, aclass) )
            print("----------------------------------------------------------------")
            
    return error
```
